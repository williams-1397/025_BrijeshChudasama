```python
# import modules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
main_data = pd.read_csv('Data_for_Transformation.csv')
plt.scatter(main_data["Age"], main_data["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
plt.hist(main_data["Salary"], bins = 10, color = "skyblue")
plt.xlabel("Salaries")
plt.show()
```

```python
fig_size = plt.figure(figsize=(7, 5))
plt.bar(main_data["Country"], main_data["Salary"], color="red")
plt.xlabel("Salaries")
plt.ylabel("Countries")
plt.title("Bar chart of country vs salary")
plt.show()
```
